```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(10,4), index = pd.date_range('2021-07-18', periods = 10), columns = ['a','b','c', 'd'])
df
```

```python
df['new'] = df['c'] + df["d"]
```

```python
df
```

```python
df['new_col_square_d'] = df['d'].apply(lambda x: x**2)
```

```python
df
```

```python
df['new_col_d_log'] = df['d'].apply(lambda x: np.log(x))
```

```python
df
```

```python
df['new_col_d_log'].fillna(0)
```

```python
df
```

```python
def func(x):
  if x>=0:
    return np.log(x)
  else:
    return x
```

```python
df['new_col_d_log_pos'] = df['d'].apply(func)
```

```python
df
```

```python
def test1(x):
  if x<0:
    return 1
  else:
    return 0
```

```python
df['new_a'] = df['a'].apply(test1)
```
